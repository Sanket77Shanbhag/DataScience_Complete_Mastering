**Array Manipulation functions**

**1. Reshaping a Numpy array**

reshape() method

```python
a = np.array([3,6,9,12])
np.reshape(a,(2,2))
```

```python
a = np.array([3,6,9,12,18,24])
print(np.reshape(a,(3,-1)))
print(np.reshape(a,(-1,3)))
print(np.reshape(a,(3,2)))
print(np.reshape(a,(2,3)))
```

**2. Flattening  a NumPy array**

flatten() or ravel()

```python
a = np.ones((2,2))
b = a.flatten()
c = a.ravel()
print(a)
print(b)
print(c)
print(a.shape)
print(b.shape)
print(c.shape)
```

**3. Transpose of a Numpy array**

```python
a = np.array([[1,2,3],[4,5,6]])
b= np.transpose(a)
print(a.shape,'\n\n',a)
print(b.shape,'\n\n',b)
```

**4. Append**

```python
a = np.array([1,2,3,4,5,6])
np.append(a,[7,8,9])
```

**5. Insert**

```python
np.insert(a,2,10)
```

**6. Delete**

```python
np.delete(a,2)
```

**7. Stacking array**

vstack() - vertical

hstack() - horizontal

```python
a = np.arange(0,5)
b = np.arange(5,10)
print(a)
print(b)
print(np.vstack((a,b)))
print(np.hstack((a,b)))
```

**8. Concatenating N dimensional array**

```python
a = np.arange(0,5).reshape(1,5)
b = np.arange(5,10).reshape(1,5)
print(a)
print(b)
print(np.concatenate((a,b),axis=0))
print(np.concatenate((a,b),axis=1))
```

```python
a = np.array([[1,6], [4,3]])
#minimum along a column print
print('Min: ',np.min(a,axis=0))
#maximum along a row print
print('Max: ',np.max(a,axis=1))
```
